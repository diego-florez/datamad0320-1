```python
import pandas as pd
import numpy as np
import sqlalchemy as db         #creates de connection with db
from getpass import getpass     #needed for the encripted password
import pymysql
```

```python
pkdb = pd.read_csv("pokemon.csv")
pkdb
```

```python
password = getpass("Mysql Password: ")
```

```python
#the connectionData var is always like that
dbName = "pokemon"
connectionData = f"mysql+pymysql://root:{password}@localhost/{dbName}"
```

```python
#necessary from now on to do anything, create tables, make queries, modify columns, etc
engine = db.create_engine(connectionData)
```

```python
#creating table in db "pkdb"
pkdb.to_sql('poke_details', con = engine, if_exists = 'append', chunksize = 1000)
```

```python
query1 = pd.read_sql_query("""
select * from poke_details
where Legendary = 1;
""", engine)
query1
```
